## **1. BASIC PLOTTING**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Getting some data
x = np.linspace(0, 5, 11) ## 11 points evenly spaced between 0 to 5
y = x**2
```

```python
## 1. Functional method of plotting something
plt.plot(x, y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
## Multiple plots on the same canvas
plt.subplot(1, 2, 1) # [number of rows, number of columns, plot number we are referring to]
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')

plt.show()
```

```python
## 2. OBJECT ORIENTED METHOD
fig = plt.figure()
# Basically we made a figure...like a canvas
# Now we added the axis where we can plot on that figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # [left, bottom, width, height]
# And now we are plotting over that axis
axes.plot(x, y)
axes.set_xlabel("X LABEL")
axes.set_ylabel("Y LABEL")
axes.set_title("Set Title")
```

```python
## Putting 2 sets of figures on one canvas
fig = plt.figure() # Making the figure object

# making the first axis
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # This is the first axes
# making other axis on the same canvas
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # This is the second axes

# plotting on axes 1
axes1.plot(x, x, 'r')
axes1.set_xlabel("axis 1 X")
axes2.set_ylabel("axis 1 y")
axes2.set_title("AXIS 1")

# plotting on axes 2
axes2.plot(x, y, 'g')
axes2.set_xlabel("axis 2 X")
axes2.set_ylabel("axis 2 y")
axes2.set_title("AXIS 2")

plt.show()
```

## **2. MORE PLOTTING**

```python
fig, axes = plt.subplots()
axes.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=5)
plt.tight_layout() # Fixes the overlap issues
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
```

```python
## PLOTTING ON THESE SUBPLOTS
# These axes objects are list of matpltotlib plots.
# So we can iterate through that and plot on each
fig, axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
  current_ax.plot(x, y)
```

```python
## If we can iterate means we can also index them
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'r')
axes[0].set_title("PLOT 1")
axes[1].plot(x, y**2, 'g')
axes[1].set_title("PLOT 2")
```

### **FIGURE SIZE AND DPI**

```python
fig = plt.figure(figsize=(3, 2), dpi=100) # dpi - dots/pixel per inch
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

### **LEGENDS**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X SQUARE')
ax.plot(x, x**3, label='X CUBE')

ax.legend() ## Calling the ax legend. We now need label object in ax.plot
```

## **3. PLOT APPEARANCE**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='blue', linewidth=5)
# color parameter can take common strings or hex or rgb
```

## **4. STATISTICAL PLOTS**

```python
## DRAWING A HISTOGRAM
data = np.random.normal(170, 10, 250)
plt.hist(data)
plt.show()
```

```python
drinks = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks.head()
```

```python
wine_data = drinks["wine_servings"]
beer_data = drinks["beer_servings"]
spirit_data = drinks["spirit_servings"]
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))  # figsize adjusts width & height

# Plot each histogram
axs[0].hist(wine_data.dropna(), bins=20, color='red', edgecolor='black')
axs[0].set_title('Wine Servings')
axs[0].set_xlabel('Servings')
axs[0].set_ylabel('Frequency')

axs[1].hist(beer_data.dropna(), bins=20, color='goldenrod', edgecolor='black')
axs[1].set_title('Beer Servings')
axs[1].set_xlabel('Servings')
axs[1].set_ylabel('Frequency')

axs[2].hist(spirit_data.dropna(), bins=20, color='purple', edgecolor='black')
axs[2].set_title('Spirit Servings')
axs[2].set_xlabel('Servings')
axs[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
